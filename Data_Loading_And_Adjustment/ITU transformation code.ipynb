```python
import pandas as pd
```

```python
data = pd.read_excel("D:\\Documents\\LUMS\\7th Semester Courses\\SPROJ\\Primary Working Data\\All ITU Data.xlsx",sheet_name="Combined Data - all countries")
```

```python
# Pivot the data as usual
pivot_df = data.pivot_table(
    index=['entityName', 'dataYear'],  
    columns='seriesName',              
    values='dataValue',                
    aggfunc='first'  # Use the first value encountered
).reset_index()

# Get the unique series units and descriptions
series_units = data.groupby('seriesName')['seriesUnits'].first()  # Get units
series_description = data.groupby('seriesName')['seriesDescription'].first()  # Assuming seriesParent is the description

# Create new DataFrames for Units and Description
units_row = pd.DataFrame([[''] * 2 + list(series_units)], columns=pivot_df.columns)
description_row = pd.DataFrame([[''] * 2 + list(series_description)], columns=pivot_df.columns)

# Set row labels to make it clear what the row is about
units_row.index = ['Units']
description_row.index = ['Description']

# Concatenate Units and Description rows with the original pivot_df
pivot_df = pd.concat([units_row, description_row, pivot_df], ignore_index=False)

# Display the updated dataframe
print(pivot_df)
```

```python
pivot_df.to_excel('All ITU Data (transformed).xlsx', index=False)
```
